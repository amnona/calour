```python
import calour as ca
%matplotlib notebook
```

```python
exp = ca.read_amplicon('data/grape-soil.biom','data/grape-soil.sample.txt',normalize=10000,filter_reads=1000, sparse=False)
```

```python
exp
```

```python
exp.cluster_features(50,inplace=True)
```

```python
exp.sample_metadata.columns
```

```python
exps=exp.sort_samples('ph')
```

```python
exps.cluster_features(inplace=True)
```

```python
exps.plot_sort('env_matter',gui='jupyter')
```

```python
sr=exps.filter_samples('env_matter','ENVO:bulk soil')
```

```python
sr.cluster_features(10,inplace=True)
```

```python
sr.plot(gui='jupyter')
```

```python
cc=sr.correlation('ph')
```

```python
cc.plot_sort('env_matter',gui='jupyter')
```
